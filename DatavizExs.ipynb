```python
import matplotlib.pyplot as plt
import numpy as np
```

Pergunta: Qual é a distribuição das alturas dos estudantes?

Tarefa: Escolha entre histograma e gráfico de barras para representar os dados. Inclua
título, rótulos dos eixos e legendas conforme apropriado.

```python
np.random.seed(2)
alturas = np.round(np.random.normal(170, 3, 300))
plt.hist(alturas, bins=int(max(alturas)-min(alturas)), edgecolor='black')
plt.title('Distribuição de Alturas')
plt.xlabel('Altura')
plt.ylabel('Quantidade')
```

Desafio – “Mapa de São Paulo”

```python
x1 = np.arange(0, 1*np.pi, 0.1)
x2 = np.arange(np.pi/2, (3*np.pi)/2, 0.1)
y1 = np.sin(x1)
y2 = np.cos(x2)

reta1x = np.arange(np.pi, 1.5*np.pi, 0.1)
y1 = np.concatenate((y1, np.zeros(len(reta1x))))

reta2x = np.arange(0, np.pi/2, 0.1)
y2 = np.concatenate((y2, np.zeros(len(reta2x))))

plt.title('Bandeira de São Paulo')
plt.axis('Off')
plt.stem(np.concatenate((x1, reta1x)), y1, linefmt='Blue', markerfmt = '.', basefmt=" ")
plt.stem(np.concatenate((x2, reta2x)), y2, linefmt='Red', markerfmt = '.', basefmt=" ")
plt.show()
```
